# 4. 다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)  
    s = '  first star   :   second star   :    third star  '  
    l = ['first star', 'second star', 'third star']


- 1) for ~ in 문을 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '
L=s.split(":")
res = []
for i in L:
    res.append(i.strip())   
print(res)
```

> - 우선 :단위로 문자열을 나누어 리스트에 넣어준다.
- 그리고 반복문을 돌려 각 단어마다 양쪽 공백을 제거하고 그 값을 res에 append내장함수를 이용하여 추가해준다.

- 2)  리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '
L=s.split(":")
res = [i.strip() for i in L ] 
print(res)
```

> - 먼저 :단위로 문자열을 나누어 리스트에 넣어준다.
- 그리고 리스트 내포를 이용하여 L에 들어있는 값마다 strip연산을 하여 좌우공백을 제거한 후 res에 리스트 형태로 저장한다.

# 5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

- [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
- [참고 2]: sum() 등의 내장함수 사용 가능함


- 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)  
        ' >>> addall([1])  
          1  
        ' >>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])  
          45



```python
def addall(L):
    sum = 0
    for i in L:
        if i>0:
            sum += i
    return sum
print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

> - addall이라는 함수는 0보다 큰 정수 리스트 변수를 인자로 받는 함수이다.
- 즉 L를 반복문을 돌리고 그 중 0보다 클 때를 sum에 더해주고 최종적으로 sum을 return해주면 된다.

```python
def addall(L):
    return sum([i for i in L if i>0])
print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

> - addall이라는 함수는 0보다 큰 정수 리스트 변수를 인자로 받는 함수이다.
- 리스트 내포를 이용하면 L안에 있는 정수 값을 하나씩 접근하여 0보다 클때인 값들을 리스트 형태로 저장할 수 있다.
- 그리고 그 리스트를 sum이라는 내장함수를 이용하여 합을 구하고 그 값을 return 한다.

- 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)  
        ' >>> addallodd([1])  
          1  
        ' >>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])  
          25



```python
def addallodd(L):
    sum = 0
    for i in L:
        if i%2 == 1:
            sum += i
    return sum
print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

> - addallodd함수는 리스트 내의 모든 홀수를 더하는 함수이다.
- 즉 인자로 받은 리스트를 반복문을 이용하여 하나씩 접근하고 그 값이 홀수인지 판별 후 홀수이면 더해나가면 된다.
- 그리고 최종적으로 더한 값을 리턴해준다.

```python
def addallodd(L):
    return sum([i for i in L if i%2==1])
print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

> - addallodd함수는 리스트 내의 모든 홀수를 더하는 함수이다.
- 이것을 리스트 내포를 이용하여 구현하려면 우선 인자로 받은 L리스트를 반복문으로 하나씩 접근하고 if문을 사용해 홀수일때의 리스트를 반환해주면 된다.
- 그리고 그 리스트를 sum이라는 내장함수를 이용하여 리턴해주면 된다.

# 7.  사전 d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}가 주어졌을 때 다음 요구사항에 맞는 코드를 제시하시오 
- [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.


- 1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
T = []
for key in d:
    T.append((key, d[key]))
T.sort()
print(T)
```

> - 우선 반복문을 이용하여 사전의 키값을 차례로 접근한다.
- 접근한 키값을 이용하여 T리스트 안에 키값과 벨류를 튜플형태로 append함수를 사용하여 추가한다.
- 그 다음 sort내장함수를 이용하여 정렬해준다.
- sort내장함수는 인자가 없으면 디폴트값으로 첫번째 인자로 오름차순 정렬한다.

- 2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
T = []
for key in d:
    T.append((key, d[key]))
T.sort(reverse = True)
print(T)
```

> - 우선 반복문을 이용하여 사전의 키값을 차례로 접근한다.
- 접근한 키값을 이용하여 T리스트 안에 키값과 벨류를 튜플형태로 append함수를 사용하여 추가한다.
- 그 다음 sort내장함수를 이용하여 정렬해준다.
- sort내장함수는 reverse값을 True로 해주면 내림차순으로 정렬이 된다.

- 3) 값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오. 

```python
def cmp(a):
    return a[1]
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
T = []
for key in d:
    T.append((key, d[key]))
T.sort(key = cmp)
print(T)
```

> - 우선 반복문을 이용하여 사전의 키값을 차례로 접근한다.
- 접근한 키값을 이용하여 T리스트 안에 키값과 벨류를 튜플형태로 append함수를 사용하여 추가한다.
- 그 다음 sort내장함수를 이용하여 정렬해준다.
- 이때 키값이 아닌 벨류값을 기준으로 정렬해 주어야 하므로 sort함수의 key값을 바꾸어 주어야 한다.
- 그러기 위해 인자로 받은 리스트의 1번 인덱스를 리턴하는 cmp라는 함수를 정의한다.
- 그리고 sort함수의 key값에 cmp를 넣으면 정렬하는 기준이 첫번째가 아닌 인덱스가 1인 두번째가 된다.
- 따라서 튜플안의 인덱스가 1인 벨류값을 기준으로 정렬을 하게 된다.
- 그리고 sort함수에 다른 인자를 넣지 않으면 디폴트로 오름차순 정렬이 된다.

- 4) 값의 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.

```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
T = []
for key in d:
    T.append((key, d[key]))
T.sort(key = cmp, reverse = True)
print(T)
```

> - 우선 반복문을 이용하여 사전의 키값을 차례로 접근한다.
- 접근한 키값을 이용하여 T리스트 안에 키값과 벨류를 튜플형태로 append함수를 사용하여 추가한다.
- 그 다음 sort내장함수를 이용하여 정렬해준다.
- 이때 키값이 아닌 벨류값을 기준으로 정렬해 주어야 하므로 sort함수의 key값을 바꾸어 주어야 한다.

# 8. [ACM-ICPC 문제 1673번] ◾치킨 쿠폰


강민이는 치킨 한 마리를 주문할 수 있는 치킨 쿠폰을 n장 가지고 있다. 이 치킨집에서는 치킨을 한 마리 주문할 때마다 도장을 하나씩 찍어 주는데, 도장을 k개 모으면 치킨 쿠폰 한 장으로 교환할 수 있다.

강민이가 지금 갖고 있는 치킨 쿠폰으로 치킨을 최대 몇 마리나 먹을 수 있는지 구하여라. 단, 치킨을 주문하기 위해서는 반드시 치킨 쿠폰을 갖고 있어야 한다.


- 입력 : 여러 줄에 걸쳐서 자연수 n과 k가 주어진다.  
- 출력 : 각 입력마다 한 줄에 정답을 출력한다.
- 제한 : 1 < k ≤ n ≤ 1,000,000,000

```python
import sys

while True:
    line = sys.stdin.readline().rstrip('\n')
    if not line:
        break
    else:
        n, k = map(int, line.split(" "))
        total = n
        a=n
        while k<=a:
            total += a//k
            a = a//k + a%k
        print(total)
```

> - 변수a는 쿠폰의 도장수라고 하고 total은 최대 치킨수이다.
- 우선 n장의 쿠폰이 있으면 n마리의 치킨을 먹을 수 있다.
- 다음으로 n마리의 치킨을 먹으면 n개의 도장을 얻을 수 있으므로 a는 n으로 초기화한다.
- 그리고 a의 수가 k보다 작을때 즉, 더이상 치킨을 먹을 수 없을때까지 아래를 반복한다.
- total에 도장수에서 k로 나눈값을 더하고 도장수는 치킨을 먹은수(a//k)와 사용하지 않은 도장수(a%k)의 합이다.
- 최대치킨의 수를 구하는 알고리즘은 이와 같고 여러 줄에 걸쳐서 자연수 n과 k가 주어진다라는 입력제시를 하기위해 인터넷을 찾아보고 위와 같이 했습니다.

      - 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '', '지향', ''] ---> ['컴퓨터공학부', '지향']

      - 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1} 

- 요구사항 5. 다음 5개의 사이트에 대한 단어 출현빈도 결과를 출력하시오.
    - http://cse.koreatech.ac.kr
    - https://www.koreatech.ac.kr
    - http://www.naver.com
    - http://www.daum.net
    - http://www.nytimes.com 



```python
import requests
import re
import string  

def call(b):  
    req = requests.get(b)
    source = req.text
    text1 = re.sub('<script.*?>.*?</script>', ' ', source, 0, re.I|re.S)
    text2 = re.sub('<style.*?>.*?</style>', ' ', text1, 0, re.I|re.S)
    text3 = re.sub('<.+?>', ' ', text2, 0, re.I|re.S)
    text4 = re.sub('-->', ' ', text3, 0, re.I|re.S)
    text5 = re.sub("[!|?|<|>|:|\[|\]|#|$|%|&|\(|\)|*|+|\-|,|.|/|;|=|@|^|_|`|{|}|~|\"|\'|\\\|/]",' ', text4, 0, re.I|re.S)
    s=text5.split()
    dic={}
    j=0
    for i in s:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1
    print(dic)

print('http://cse.koreatech.ac.kr')    
call('http://cse.koreatech.ac.kr')
print()

print('https://www.koreatech.ac.kr')
call('https://www.koreatech.ac.kr')
print()

print('http://www.naver.com')
call('http://www.naver.com')
print()

print('http://www.daum.net')
call('http://www.daum.net')
print()

print('http://www.nytimes.com')
call('http://www.nytimes.com')
```
