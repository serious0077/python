* 2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.

```python
def cmp(a):
    return a[4]

L = []
with open ('s2.txt', 'w') as f:
    f.write("비었습니다")
with open ('s2.txt', 'r') as f:
    print(f.read()) 
    
print()


with open('s.txt', 'r') as f:
    L = f.readlines()
    L.sort(key = cmp)
with open('s2.txt', 'w') as f:
    f.write(''.join(L))
with open ('s2.txt', 'r') as f:
    print(f.read())
```

> * 우선 cmp라는 함수를 선언하는데 인자로 받은 것의 인덱스4를 반환한다.

* 그 다음 sort내장함수를 사용하여 정렬하는데 sort의 인자값중 하나인 key를 cmp로 지정해준다.
* 그러면 각 문자열의 4번 인덱스를 기준으로 정렬이 된다.
* 그리고 이 리스트를 join내장함수를 이용하여 하나의 문자열로 이어주고 s2.txt에 저장한다.

# 3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

* 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.

>range(stop)  
range(start, stop[, step])

* 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 

* 3) frange 사용 예
    * 3-1) frange(0.5)
        * [0.0, 0.1, 0.2, 0.3, 0.4]

    * 3-2) frange(1.0, 2.0)
         * [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]

    * 3-3) frange(2.2, 4.0, 0.5)
         * [2.2, 2.7, 3.2, 3.7]



```python
def frange(stop, start = 0.0, step = 0.1):
    list = []
    if(start > stop):
        start, stop = stop, start
    curr = start
    while True:
        if curr < stop:
            list.append(round(curr, 3))
            curr = curr + step
        else:
            break
    return print(list)
```

```python
frange(0.5)
frange(1.0, 2.0)
frange(2.2,4.0,0.5)
```

> * 먼저 frange라는 함수를 정의한다.
* 함수의 인자는 stop, start, step의 3개이고 step과 start는 각각 0.1과 0.0으로 기본 인수 값을 지정해준다.
* 그리고 함수를 호출할 때 인자를 하나, 둘, 셋까지 각각 따로 줄수있다.
* 이때 기본 값을 지닌 인수 뒤에 일반적인 인수가 올 수 없으므로 일반적인 인수인 stop이 가장 앞에 와야되고 이 때문에 인자가 두개 이상 들어올 경우를 생각하여 우선 start에 stop값을 그리고 stop값에 start값을 받는다.
* 그리고 이 두 값을 swap해준다.
* 그리고 curr는 start에서 시작해서 stop보다 같거나 커질때까지 step만큼 증가하며 list에 저장된다.
* 마지막으로 step마다 증가된 curr값이 저장된 list를 반환한다.

# 4. 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오.

* sum() 사용 예 
    * sum()
        * 0

    * sum(1, 2)
        * 3

    * sum(1, 2, 3, 4, 5)
        * 15

    * sum(1, 5, 7, 2, -10)
        * 5

```python
def sum(*args, a=0):
    sum = 0
    for i in args:
        sum = sum + i
    return print(sum)
```

```python
sum()
sum(1,2)
sum(1,2,3,4,5)
sum(1,5,7,2,-10)
```

# 6. 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
* [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다. 


```python
def myfact(n):
    if n < 0:
        print("음의 정수를 입력하셨습니다.")
        return
    if n == 0:
        return 1
    else:
        return myfact(n-1)*n
```

```python
for i in range(100):
    print(myfact(i))
```

> * 우선 n이 음수일 때 '음의 정수를 입력하셨습니다.'라고 사용자에게 알려준다.
* 그리고 n이 0일때 1을 반환한다.
* 마지막으로 그 외에는 n과 myfact(n-1)을 곱한다.
* 즉  myfact(n) *  myfact(n-1) *  myfact(n-2) * ..... *  myfact(2) *  myfact(1) * myfact(0)가 된다.
* 위의 식을 풀어보면 n * (n-1) * (n-2) *...* 2 * 1 * 1이 된다.
* 이것은 팩토리얼 계산법과 같다.
* 즉 함수 자기자신을 호출하는 재귀함수를 이용하여 팩토리얼계산이 가능하다.

* 5) 위 문제에서 저장한 모든 pickle 파일들을 객체로 다시 로드하여 본인이 저장하여 분석한 사이트들 각각에 대해서 가장 많이 출현한 단어 3개씩를 뽑아 제시하시오. 
    * 반드시 pickle 모듈로 저장한 5개 이상의 pickle 파일들을 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함


```python
def load(load):
    f = open(load, 'rb')                         
    frequency = pickle.load(f)
    f.close()
    return frequency

def frequency(fre):
    dic = {}
    
    a = list(fre.values())
    a.sort(reverse = True)

    count = 0
    for j in range(3):
        for i in fre:
            if count==3:
                break
            if fre[i] == a[j]:
                count = count + 1
                dic[i] = a[j]
    return dic


naver_words_frequency = load('www.naver.com.words_frequency.pickle.txt')
google_words_frequency = load('www.google.com.words_frequency.pickle.txt')
youtube_words_frequency = load('www.youtube.com.words_frequency.pickle.txt')
cnn_words_frequency = load('www.cnn.com.words_frequency.pickle.txt')
kbs_words_frequency = load('www.kbs.kr.words_frequency.pickle.txt')

print('http://www.naver.com', ' : ', frequency(naver_words_frequency))
print('http://www.google.com', ' : ', frequency(google_words_frequency))
print('http://www.youtube.com', ' : ', frequency(youtube_words_frequency))
print('http://www.cnn.com', ' : ', frequency(cnn_words_frequency))
print('http://www.kbs.kr', ' : ', frequency(kbs_words_frequency))
```

* 그 다음 리스트는 정렬이 가능하므로 sort(reverse = True)를 사용하여 내림차순으로 정렬하여 변수a에 저장한다.
* 이렇게 하여 출현 빈도를 가지고 있는 변수a에서 인덱스가 0이 가장 출현횟수가 큰 것이 된다.
* 따라서 단어를 출현빈도가 가장 높은 3개의 단어를 구하는 것이므로 a의 0에서 2까지의 범위만 사용하면 된다.
